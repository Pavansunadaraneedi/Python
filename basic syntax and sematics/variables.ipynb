Variable 
--------
 A variable in Python is a name that refers to a value stored in memory. It allows you to store, modify, and retrieve data in your program.

```python
a=100
```

```python
## Declaring and Assigning Variable

age = 23
height = 5.6
name = "Pavan"
is_student = True

## Printing the variables

print("age :",age)
print("Height :",height)
print("Name :",name)
```

```python
## Naming Conventions
## Variable names should be descriptive
## They must start with a letter or an '_' and contains letter, number and underscores
## Variables names case sensitive

# Valid variables names 

first_name = "Pavan"
last_name = "Sundaraneedi"
```

```python
#Invalid variable names

# 2age = 23
#first-name = "Krish"
#@name = "krish"
```

```python
## case sensitivity 
name="pavan"
Name="pavone"
```

```python
## Understanding the variable types
## Python is dynamically typed, type of a variable is determined at runtime

age = 23 #int
height = 5.7 #float
name = "krish" #str
is_student =True #bool

print(type(age))
print(type(height))
print(type(name))
print(type(is_student))
```

```python
## Dynamic Typing
## Python allows the type of a variable to change as the program executes

var=10 #int
print(var,type(var))

var="Hello"
print(var,type(var))

var=3.14
print(var,type(var))
```
