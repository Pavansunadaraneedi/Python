**Introduction to Functions**

Definition:

A function is block of code that performs a specific task. functiond help in organizing code, reusing code, and improving readability.

```python
## Syntax
def function_name(parameters):
   """Docstring"""
   #function body
   return expression
```

```python
## Why functions?
num=24
if num%2==0:
    print("Number is even")
else:
    print("NUmber is odd")
```

```python
def even_or_odd(num):
   """This function finds even or odd"""
   if num%2==0:
      print("The number is even")
   else:
      print("The number is odd")
```

```python
# Call this Function
even_or_odd(24)
even_or_odd(11)
even_or_odd(9)
```

```python
## function with multiple parameters

def add(a,b):
    return a+b

result = add(2,4)
print(result)
```

```python
## Default parameters

def greet(name):
    print(f"Hello {name},  Welcome to my team")

greet("Pavan")
```

```python
def greet(name="Guest"):
    print(f"Hello {name}, Welcome to my team")

greet()
```

```python
### Variable Length Arguments
### Positional And keywords arguments

def print_numbers(*Pavan):
    for number in Pavan:
        print(number)
```

```python
print_numbers(1,2,3,4,5,6,7,8,"Pavan")
```

```python
## Positional arguments
def print_numbers(*args):
    for number in args:
        print(number)
```

```python
print_numbers(1,2,3,4,5,6,7,8,"Pavan")
```

```python
### Keyword Arguments

def print_details(**kwargs):
    for key,value in kwargs.items():
        print(f"{key}:{value}")
    
```

```python
print_details(name="Pavan",age="23",country="India")
```

```python
def print_details(*args,**kwargs):
    for val in args:
        print(f"Positional argument :{val}")

    for key,value in kwargs.items():
        print(f"{key}:{value}")
```

```python
print_details(1,2,3,4,"Pavan",name="Pavan",age="23",country="India")
```
