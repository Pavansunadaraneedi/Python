### **Q1. What are data types in Python?**

**A:** Data types define the type of value a variable can store, such as numbers, text, lists, etc.

---

### **Q2. What is an `int` in Python?**

**A:** `int` represents whole numbers without decimals.
Example: `10`, `-5`.

---

### **Q3. What is a `float`?**

**A:** `float` represents decimal or floating-point numbers.
Example: `3.14`.

---

### **Q4. What is a `string` (`str`)?**

**A:** A string is a sequence of characters enclosed in quotes.
Example: `"Hello"`.

---

### **Q5. What is a `bool`?**

**A:** Boolean data type represents `True` or `False`.

---

### **Q6. What is a `list`?**

**A:** List is an ordered, mutable collection that allows duplicates.
Example: `[1, 2, 3]`.

---

### **Q7. What is a `tuple`?**

**A:** Tuple is an ordered, immutable collection.
Example: `(1, 2, 3)`.

---

### **Q8. What is a `set`?**

**A:** Set is an unordered collection of unique elements.
Example: `{1, 2, 3}`.

---

### **Q9. What is a `dictionary` (`dict`)?**

**A:** Dictionary stores data in keyâ€“value pairs.
Example: `{"name": "John"}`.

---

### **Q10. What is `None`?**

**A:** `None` represents the absence of a value (null).

### **Q20. What is OOP in Python?**

**A:** Object-Oriented Programming is a method of structuring code using classes and objects.

---

### **Q21. What is a class?**

**A:** A class is a blueprint for creating objects.

---

### **Q22. What is an object?**

**A:** An object is an instance of a class, containing data and methods.

---

### **Q23. What is a constructor (`__init__`)?**

**A:** A special method that automatically runs when an object is created.

---

### **Q24. What is encapsulation?**

**A:** Wrapping data and methods into a single unit and controlling access.
(Using private `__var`, protected `_var` variables.)

---

### **Q25. What is inheritance?**

**A:** A child class acquires properties and methods of a parent class.

---

### **Q26. What is polymorphism?**

**A:** Same function/method name showing different behaviors depending on the object.

---

### **Q27. What is method overriding?**

**A:** When a child class provides its own version of a method already defined in the parent class.

---

### **Q30. What is `super()`?**

**A:** A function used to call methods/constructors of the parent class.
