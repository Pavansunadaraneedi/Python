**3.Remove Duplicate in a List**

```python
def remove_duplicates(lst):
    st = set(lst)
    lst=list(st)
    return lst
remove_duplicates([1, 2, 2, 3, 4, 4, 5])
```

**4. Check if all Elements in a List are Unique**

```python
def check_unique(lst):
    seen = set()
    for num in lst:
        if num in seen:
            return False  # Duplicate found
        seen.add(num)
    return True  #
    
check_unique([1, 2, 3, 3, 4, 5])
```

**5.Reverse a List**

```python
def reverse_list(lst):
    lst.reverse()
    
    return lst
reverse_list([1, 2, 3, 4, 5])
```

**6.Number of odd and even Elements in a List**

```python
def count_even_odd(lst):
    z=0
    y=0
    for x in lst:
        if x%2==0:
            z+=1
        else:
            y+=1
    return z,y
count_even_odd([1, 2, 3, 4, 5])
```

**7.Check if List is Subset of Another List**

```python

def is_subset(lst1, lst2):
    lst1 = set(lst1)  # Convert lst1 to set
    lst2 = set(lst2)  # Convert lst2 to set
    
    x = lst1.issubset(lst2)  # Correct method name: issubset
    return x

# Test the function
print(is_subset([1, 2, 3], [1, 2, 3, 4, 5]))  # Output: True
```

**8.Maximum Difference between 2 Consecutive elements in a List**

```python
def max_consecutive_difference(lst):
    # Initialize max_difference to a very small number
    max_diff = 0
    
    # Loop through the list and calculate the difference between consecutive elements
    for i in range(1, len(lst)):
        diff = abs(lst[i] - lst[i - 1])  # Difference between consecutive elements
        max_diff = max(max_diff, diff)  # Update max_diff if current diff is larger
    
    return max_diff

# Test the function
print(max_consecutive_difference([1, 7, 3, 10, 5]))  # Output: 7

```
