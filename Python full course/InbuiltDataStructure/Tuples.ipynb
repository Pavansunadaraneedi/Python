Introduction to the tuples

Tuples:

Tuples are ordered collection of items that are immutavle. They are similar to list , but their immutability makes them different.

```python
## creating a tuple

empty_tuple=()
print(empty_tuple)
print(type(empty_tuple))
```

```python
lst=list()
print(type(lst))
tpl=tuple()
print(type(tpl))
```

```python
numbers=tuple([1,2,3,4,5,6])
numbers
```

```python
list((1,2,3,4,5,6))
```

```python
mixed_tuple=(1,"hello world", 3.14, True)
print(mixed_tuple)
```

```python
## Accessing Tuple Elements
numbers
```

```python
print(numbers[2])
print(numbers[-1])
```

```python
numbers[0:4]
```

```python
numbers[::-1]
```

```python
## Tuple operation

concatenation_tuple = tuple(numbers) + mixed_tuple
print(concatenation_tuple)
```

```python
mixed_tuple * 3
```

```python
numbers * 3
```

```python
## Immutable nature of Tuples
## Tuple are immutable, meaning their elements canno be changed once assigned.

lst=[1,2,3,4,5]
print(lst)

lst[1] = "Pavan"
print(lst)
```

```python
numbers = (1,2,3,4,5,6)
numbers[1]="Pavan"
```

```python
## Tuple Methods
print(numbers.count(1)) #count
print(numbers.index(3)) #index
```

Packing and Unpacking tuple

```python
## Packing
packed_tuple=1,"Hello",3.14
print(packed_tuple)
```

```python
## Unpacking tuple
a,b,c = packed_tuple
print(a)
print(b)
print(c)
```

```python
## Unpacking with *
numbers=(1,2,3,4,5,6)
first,*middle,last=numbers
print(first)
print(middle)
print(last)
```

Nested tuple

```python
## Nested list
lst=[[1,2,3,4],[6,7,8,9],[1,"Hello",3.14,"c"]]
lst[0][0:3]
```

```python
lst=[[1,2,3,4],[6,7,8,9],(1,"Hello",3.14,"c")]
lst[2][0:3]
```

```python
nested_tuple = ((1,2,3),("a","b","c"),(True, False))

## access the element inside a tuple
print(nested_tuple[0])
print(nested_tuple[1][2])
```

```python
## iterating over nested tuples
for sub_tuple in nested_tuple:
    for item in sub_tuple:
        print(item,end=" ")
    print()
```
