**Dictionaries**

Dictionaries are unordered collection of items. They store data in key-values pairs. Keys must be unique and immutable
(Eg: Strings, Numbers, or Tuple), While values can be of any type. 

```python
## Creating Dictionaries
empty_dict={}
print(type(empty_dict))
```

```python
empty_dict=dict()
empty_dict
```

```python
student={"name":"Pavan","age":23,"grade":24}
print(student)
print(type(student))
```

```python
## single key is always used
student={"name":"Pavan","age":23,"name":24}
print(student)
```

```python
## Accessing Dictionary element
student={"name":"Pavan","age":23,"grade":"A"}
print(student)
```

```python
## Accesssing Dictionary elements
print(student["grade"])
print(student["age"])

## Accessing using get() method
print(student.get('grade'))
print(student.get('last_name'))
print(student.get('last_name',"Not Available"))
```

**Modifying Dictionary Elements**

```python
## Dictionary are mutable, so you can add, update or delete elements
print(student)
```

```python
student["age"]=24 # Update value for the key
print(student)
student["Address"]="India" # Added a new key and value
print(student)
```

```python
del student['grade'] ## Delete key and value pair
print(student)
```

**Dictionary Methods**

```python
keys=student.keys() # get all the keys
print(keys)
values=student.values() # get all the values
print(values)

items=student.items() # get all key value pairs
print(items)
```

```python
## Shallow copy
student_copy=student
print(student)
print(student_copy)
```

```python
student["name"]="Pavan2"
print(student)
print(student_copy)
```

```python
student_copy1=student.copy() ## Shallow copy
print(student_copy1)
print(student)
```

```python
student["name"]="Pavan3"
print(student_copy1)
print(student)
```

```python
### Iterating Over dictionaries
## You can use loops to iterate over dictonaries, keys, values, or items

## Iterating over keys
for keys in student.keys():
 print(keys)
```

```python
## iterate over values 
for values in student.values():
 print(values)
```

```python
## Iterate over key value pairs
for key,values in student.items():
    print(f"{key}:{values}")
```

```python
## Nested dictionaries
students={
    "student1":{"name":"Pavan","age":23},
    "student2":{"name":"Ammu","age":24}
}
print(students)
```

```python
## Access Nested dictionaries elements
print(students["student2"]["name"])
print(students["student2"]["age"])
```

```python
students.items()
```

```python
## Iterating over Nested Dictionaries
for student_id,student_info in students.items():
    print(f"{student_id}:{student_info}")
    for key, values in student_info.items():
        print(f"{key}:{values}")
        
```

```python
## Dictionary comprehension
squares={x:x**2 for x in range(5)}
print(squares)
```

```python
## Conditional Dictionary comprehension
evens={x:x**2 for x in range(10) 
       if x%2==0
       }
print(evens)
```

```python
## Practical Examples

## Use a dictionary tp count the frequency of elements in list

numbers=[1,2,2,3,3,3,4,4,4,4]
frequency={}

for number in numbers:
    if number in frequency:
        frequency[number]+=1
    else:
        frequency[number]=1
print(frequency)
```

```python
## Merge two dictionaries into one
dict1={"a":1,"b":2}
dict2={"b":3, "c":4}
merged_dict={**dict1,**dict2}
print(merged_dict)
```
