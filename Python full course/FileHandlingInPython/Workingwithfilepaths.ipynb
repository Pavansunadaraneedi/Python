**Working with File Paths in Python**

Python provides tools like os.path and pathlib to work with file and folder locations. These tools help you create paths, join folders, check whether files exist, and read or write files safely. pathlib offers a cleaner, object-oriented way to handle paths across different operating systems. Understanding absolute/relative paths, directory creation, and error handling makes file operations more reliable and portable.

```python
import os 
cwd=os.getcwd()
print(f"Current working directory is {cwd}")
```

```python
## create a new directory
new_directory="package"
if not os.path.exists(new_directory):
	os.mkdir(new_directory)
	print(f"Directory '{new_directory}' created")
else:
	print(f"Directory '{new_directory}' already exists")
```

```python

## Listing Files And Directories
items=os.listdir('.')
print(items)

```

```python
## Joining paths

dir_name="floder"
file_name="file.txt"
full_path=os.path.join(os.getcwd(),dir_name,file_name)
print(full_path)
```

```python
path='example.txt'
if os.path.exists(path):
    print(f"The path '{path}' exists")
else:
    print(f"The path '{path}' does not exits")
```

```python
## Checking if a path is a File or directory
import os

path ='example.txt'
if os.path.isfile(path):
    print(f"The path '{path}' is a file")
elif os.path.isdir(path):
    print(f"The path '{path}' is a directory.")
else:
    print(f"The path '{path}' is neither a file nor a directory.")
          
```

```python
## Getting the absolute path 
import os
relative_path='example.txt'
absolute_path=os.path.abspath(relative_path)
print(absolute_path)
```
