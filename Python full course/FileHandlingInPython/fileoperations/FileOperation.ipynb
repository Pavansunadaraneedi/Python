**File Operation-Read And Write Files**

File handling is a crucial part of any programming language. Python provides built-in functions and methods to read from and write to files, both text and binary. This lesson will cover the basics of file handling, including reading and writing text file and binary files.

```python
### Read a whole File

with open('example.txt','r') as file:
    content=file.read()
    print(content)
```

```python
### Read a file line by line 
with open('example.txt','r') as file:
    for line in file:
        print(line.strip()) ## strip() removes the newline character
```

```python
## Writting a file(Overwritting)

with open('example.txt','w') as file:
    file.write("Hello World!\n")
    file.write("This is a New Line.")
```

```python
## Write a file without overwritting

with open('example.txt','a') as file:
    file.write("Append operation taking place!")
```

```python
### Writting a list of line of line to a file 
lines=['First line ','Second line ', 'Third line ']
with open('example.txt','a') as file:
    file.writelines(lines)
```

```python
## Binary Files

#Write to a binary file
data=b'\x00\x01\x02\x04'
with open('example.bin', 'wb') as file:
    file.write(data)
```

```python
# Reading a binary file
with open('example.bin', 'rb') as file:
    content = file.read()
    print(content)
```

```python
## Read the content from source text file and write to a destination text file
# Copying a text file
with open('example.txt', 'r') as source_file:
    content = source_file.read()

with open('destination.txt', 'w') as destination_file:
    destination_file.write(content)
```

```python
# Read a text file and count the number of lines, words and character.
# Counting lines, words, and characters in a text file
def count_text_file(file_path):
    with open(file_path,'r') as file:
        lines=file.readlines()
        line_count= len(lines)
        word_count=sum(len(line.split()) for line in lines)
        char_count = sum(len(line) for line in lines)
    return line_count, word_count, char_count

file_path = 'example.txt'
lines, words, characters = count_text_file(file_path)
print(f'Lines: {lines}, Words: {words}, Characters: {characters}')
```

The w+ mode in Python is used to open a file for both reading and writing. If
the file does not exist, it will be created. If the file exists its content truncated.

i.e the file is overwritten.

```python
### Writting and Then Reading a file

with open('example.txt','w+') as file:
    file.write("Hey Hello world\n")
    file.write("This is a new line\n")

    ## Move the file cursor to the beginning
    file.seek(0)

    ## Read the conent of the file
    content=file.read()
    print(content)

```
